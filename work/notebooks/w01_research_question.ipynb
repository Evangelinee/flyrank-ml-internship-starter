## 2. The question: decision, action, cost of a wrong call

**Search question:** Which pseudonymized content pages should an editor review first for refresh, expansion, protection, or monitoring?

**Unit of analysis:** One pseudonymized content page.

**Decision and output:** The work will produce a ranked review queue with a priority score, suggested action, reason codes, and a confidence label. A content editor can inspect the highest-ranked pages first and decide whether to refresh, expand, protect, or monitor each page.

**Cost of a wrong call:** A false positive wastes limited editor time and an unnecessary change could disrupt a page that was performing adequately. A false negative leaves a genuinely declining opportunity unreviewed. The queue therefore supports human review; it does not automate publication decisions.

**Why data or ML can help:** Impressions, position, CTR, age, freshness, and engagement may interact in ways that one fixed threshold cannot cover. I will begin with a transparent rule and use a model only if it improves Precision@K under client-grouped validation. The goal is a better review decision, not merely to train a model.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
declining = df["trend_direction"].astype(str).str.lower().eq("down")
stale_visible = (
    (df["days_since_last_update"] >= 180)
    & (df["impressions_90d"] >= 500)
)

print("Decision grain: one row = one pseudonymized content page")
print("Action capacity example: review the top 20 or top 50 pages")
print("Baseline: stale >= 180 days AND impressions_90d >= 500")
```

## 3. Quick look at the data (2-3 real numbers)

The starter slice contains **30,000 pages**. It records **16,262 declining pages (54.2%)**, so decline is common enough to make prioritization relevant. A strict stale-and-visible baseline selects only **17 pages**, and **16 of those 17 (94.1%)** are declining.

This is promising but too narrow for a complete workflow: a ranked method may preserve useful precision while expanding the queue to the team's actual review capacity. These are observed starter-sample counts, not evidence that refreshing a page will cause recovery.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
total_pages = len(df)
declining_count = int(declining.sum())
declining_rate = declining.mean()
baseline_count = int(stale_visible.sum())
baseline_declining = int((stale_visible & declining).sum())
baseline_rate = declining[stale_visible].mean()

print(f"Total pages: {total_pages:,}")
print(f"Observed declining pages: {declining_count:,} ({declining_rate:.1%})")
print(f"Strict stale-visible candidates: {baseline_count:,}")
print(
    f"Declining among strict candidates: {baseline_declining:,} "
    f"of {baseline_count:,} ({baseline_rate:.1%})"
)
```
